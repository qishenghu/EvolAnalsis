```python
from agentevolver.client.env_client import EnvClient
```

```python
client = EnvClient(base_url="http://127.0.0.1:8081")
```

```python
# 1）获取 seed task 列表
task_ids = client.get_env_profile(env_type="alfworld", split="train")
print("Num train tasks:", len(task_ids), "first 5:", task_ids[:5])  # 预期 2420
```

```python
# 2）创建一个环境实例
tid = task_ids[225]
# 2）显式指定一个 instance_id（和训练里的做法一致）
instance_id = None

init = client.create_instance(
    env_type="alfworld",
    task_id=tid,
    instance_id=instance_id,          # ⭐ 显式传入
    params={"world_type": "Text"},
)
```

```python
print(init['state'][1]['content'])
```

```python
print("Init keys:", init.keys())
print("First obs:", init["state"][1]["content"][:200], "...")
print("Sample actions:", init["available_actions"][:5])
```

```python
print(init["state"][1]["content"])
```

```python
instance_id = init['instance_id']
```

```python
instance_id
```

```python
# 3）执行一步
action = {"content": "go to drawer 1"}  # 或者 init["available_actions"][0]
step_out = client.step(instance_id=instance_id, action=action)
print("Reward:", step_out["reward"], "Done:", step_out["is_terminated"])
print("Next obs:", step_out["state"][-1]["content"][:200], "...")
```

```python
step_out
```

```python
# 4）评估（可选）
score = client.evaluate(instance_id=instance_id)
print("Eval score:", score)
```

```python
# 5）释放实例
ok = client.release_instance(instance_id)
print("Released:", ok)
```
